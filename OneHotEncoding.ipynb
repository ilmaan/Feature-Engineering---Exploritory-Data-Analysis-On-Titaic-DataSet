ONE HOT ENCODING

One Hot Encoding

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('test.csv',usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
data.head()
```

```python
for col in data.columns:
  print(col,':',len(data[col].unique()),'labels')
```

```python
pd.get_dummies(data, drop_first=True).shape
```

```python
data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
for label in top10:
  data[label] = np.where(data['X2']==label,1,0)

data[['X2']+top10].head(40)
```

```python
def onehot(df,var,tox):
  for label in tox:
    df[var+'_'+label] = np.where(data[var]==label, 1, 0)

onehot(data,'X2',top10)
```
